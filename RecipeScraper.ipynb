The purpose of this project is to take a recipe query, search google for the top 5 results, scrape the recipe data from the website, clean the recipe data, and provide a side-by-side view of the recipes for easy comparison of ingredients and procedures.

```python
from googlesearch import search 

# enter your recipe query
query = "recipe for chicken shawarma"

# create an empty list to add the google results to
results = []
i = 0

# return google results and append to list.
for item in search(query, tld="co.in", num=5, stop=5, pause=2):
    results.append(item)
    print(item) 
    i = i + 1
```

```python
import requests
from bs4 import BeautifulSoup

# change this number to look at each of the different recipes returned from the search
page_link = results[2]

# access the website
page_response = requests.get(page_link, timeout=5)

# scrape the data into our beautiful soup
soup = BeautifulSoup(page_response.content, "html.parser")

# print formatted HTML code
# print(page_content.prettify())
```

```python
# print the text of the page with lots of native whitespace
# print(soup.get_text())

# strip whitespace from the page text, append strings to a list 
total_string = []
for string in soup.stripped_strings:
    print(repr(string))
    total_string.append(string)
```

```python
# This section is not robust - 
# We assume all recipes begin with "Ingredients"
# We drop all the useless text that comes before this word

i = 0
search = total_string[i]

while "Ingredients" not in search:
    total_string.remove(total_string[0])
    print(search)
    i = i + 1
    search = total_string[i]
```
